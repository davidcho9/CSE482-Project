```python
import pandas as pd

corn_path = 'datasets/corn.csv'
gold_path = 'datasets/gold.csv'
oil_path = 'datasets/oil.csv'
dow_path = 'datasets/dow.csv'
twitter_path = 'datasets/twitter.csv'
micro_path = 'datasets/MSFT.csv'

corn = pd.read_csv(corn_path)
gold = pd.read_csv(gold_path)
oil = pd.read_csv(oil_path)
dow = pd.read_csv(dow_path)
twitter = pd.read_csv(twitter_path)
microsoft = pd.read_csv(micro_path)

#REMOVE MISSING VALUES FROM OIL DATASET
missing_index = oil[oil['DCOILWTICO'] == '.'].index
oil = oil.drop(missing_index)

#REMOVE MISSING VALUES GOLD DATASET
missing_index = gold[gold['GOLDAMGBD228NLBM'] == '.'].index
gold = gold.drop(missing_index)

corn = corn[::-1]
corn = corn.reset_index()
del corn['index']
twitter
```

```python
import re

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# for i,v in corn.iterrows():
#     corn['Date'] = 's'

def getMonth(target, l_months):
    
    for i,v in enumerate(l_months):
        if v == target:
            if (i+1) < 10:
                return '0' + str(i+1) 
            else:
                return str(i+1)
            
def reformatDate(old_date, l_months):
    

    new_date = re.split('\s', old_date)
    new_date[1] = new_date[1].strip(',')
    
    year = new_date[2]
    month = getMonth(new_date[0], l_months)
    day = new_date[1]
    
    d = year + '-' + month + '-' + day
    
    return d

corn['Date'] = corn.apply(lambda row : reformatDate(row['Date'], months), axis=1) 
corn.shape
```

```python

def syncDates(df1, df2, k1, k2):
    
    for i,v in enumerate(df1[k1].values):
        if (v not in df2[k2].values):
            missing_index = df1[df1[k1] == v].index
            df1 = df1.drop(missing_index) 

    for i,v in enumerate(df2[k2].values):
        if (v not in df1[k1].values):
            missing_index = df2[df2[k2] == v].index
            df2 = df2.drop(missing_index) 
    return df1, df2
            
print(gold.shape)
gold, oil = syncDates(gold, oil, 'DATE', 'DATE')
#gold, corn = syncDates(gold, corn, 'DATE', 'Date')
gold, dow = syncDates(gold, dow, 'DATE', 'Date')
gold, twitter = syncDates(gold, twitter, 'DATE', 'Date')
twitter, oil = syncDates(twitter, oil, 'Date', 'DATE')
twitter, dow = syncDates(twitter, dow, 'Date', 'Date')
twitter, corn = syncDates(twitter, corn, 'Date', 'Date')
twitter, microsoft = syncDates(twitter, microsoft, 'Date', 'Date')

print('Gold: ', gold.shape)
print('Oil: ', oil.shape)
print('Corn: ', corn.shape)
print('Dow: ', dow.shape)
print('Twitter: ', twitter.shape)
```
